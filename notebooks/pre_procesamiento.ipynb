```python
import pandas as pd
import numpy as np
import os
```

```python

df = pd.read_csv(os.path.join('mvp_playstore\data\googleplaystore.csv'), encoding='latin-1')
# Limpiar columna Installs
df['Installs'] = df['Installs'].str.replace('[+,]', '', regex=True)
df['Installs'] = pd.to_numeric(df['Installs'], errors='coerce')
```

```python
# Limpiar columna Size (convertir M y k a números)
def parse_size(s):
    if pd.isna(s) or s == 'Varies with device':
        return np.nan
    s = str(s)
    if 'M' in s:
        return float(s.replace('M', '')) * 1024
    elif 'k' in s:
        return float(s.replace('k', ''))
    return np.nan

df['Size'] = df['Size'].apply(parse_size)
```

```python
# Limpiar Price
df['Price'] = df['Price'].str.replace('$', '', regex=False)
df['Price'] = pd.to_numeric(df['Price'], errors='coerce').fillna(0)

# Limpiar Rating
df['Rating'] = pd.to_numeric(df['Rating'], errors='coerce')

# Variable objetivo: Éxito = 1 si instalaciones > 100,000
df['Exito'] = (df['Installs'] >= 100000).astype(int)

# Eliminar nulos
df.dropna(subset=['Rating', 'Size', 'Installs'], inplace=True)
```

```python
# Codificar variables categóricas
df['Type_num'] = (df['Type'] == 'Free').astype(int)
df['Category_num'] = pd.factorize(df['Category'])[0]
df['ContentRating_num'] = pd.factorize(df['Content Rating'])[0]

df.to_csv('mvp_playstore\data\googleplaystore_preprocesado.csv', index=False, encoding='utf-8')
```
